```python
import pandas as pd
```

```python
melb_df = pd.read_csv('melb_data_fe.csv')
```

```python
melb_data = melb_df.copy()
```

```python
melb_data.head()
```

```python
melb_data.columns
```

```python
melb_data['Quarter'] = pd.to_datetime(melb_data['Date']).dt.quarter
```

```python
second_most_pop_quater = melb_data['Quarter'].value_counts().iloc[1]
second_most_pop_quater
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_values = 150
for col in melb_data.columns:
    if melb_data[col].nunique() < max_unique_values and col not in cols_to_exclude:
        melb_data[col] = melb_data[col].astype('category')
```

```python
melb_data.sort_values(by='AreaRatio', ascending=False, ignore_index=True).iloc[1558]
```

```python
melb_data_townhouse = melb_data[(melb_data['Type'] == 'townhouse') & (melb_data['Rooms'] > 2)]
melb_data_townhouse.sort_values(by=['Rooms', 'MeanRoomsSquare'], ascending=[True, False], ignore_index=True).iloc[18]
```
